## Предобработка данных

Столбец first_day_exposition имеет формат object, переведем его в формат даты 

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'])
```

Данные количества комнат и  цены приведем к целым числам

```python
data['rooms'] = data['rooms'].astype('int64')
data['last_price'] = data['last_price'].astype('int64')
```

### Обработка пропусков


Пропущенные значение в is_apartment установим на False, скорее всего это некорректная выгрузка или пользователь не указал что его квартира не является апартаментами

Заполним пропуски - установив False в местах пропусков,
а затем изменим тип на bool 

```python
data['is_apartment'] = data['is_apartment'].fillna(value = False)
data['is_apartment'] = data['is_apartment'].astype('bool')
```

Заполним пропущенные значения в цифровых данных на минус 1 и изменим тип данных на целый  

```python
data['days_exposition']= data['days_exposition'].fillna(-1).astype('int64', errors='ignore')
data['parks_around3000']= data['parks_around3000'].fillna(-1).astype('int64', errors='ignore')
data['ponds_around3000']= data['ponds_around3000'].fillna(-1).astype('int64', errors='ignore')
data['airports_nearest']= data['airports_nearest'].fillna(-1).astype('int64', errors='ignore')
data['parks_nearest']= data['parks_nearest'].fillna(-1).astype('int64', errors='ignore')
data['ponds_nearest']= data['ponds_nearest'].fillna(-1).astype('int64', errors='ignore')
```

В столбце количества балконов есть пропуски - скорее всего пропуск - это отсутствие балкона. Заменим пропуски нулями

```python
data['balcony'] = data['balcony'].fillna(0)
```

Удалим данные где не указаны в характеристике квартиры - город ее расположения, потому что эти данные составляют небольшой объем 0.2% и их наличие обязательно для дальнейшего анализа.

```python
data = data.dropna(subset = ['locality_name'])
```

Удалим данные где не указано значение 'floors_total' так как они занимают 0.3 % выборки
а заменить значение на среднее или приравнять значения этажности дома к этажу квартиры 
изменит данные - так как эти квартиры автоматически станут квартирами на последнем этаже

```python
data = data.dropna(subset = ['floors_total'])
```

Изменим формат данных на целочисленный

```python
data['floors_total'] = data['floors_total'].astype('int64')
```

Пропущенные данные из столбцов 'living_area' и 'kitchen_area' оставим без изменения, удалять и усреднять не будем, чтобы сохранить данные, в будущем они могут пригодиться.

```python
data.isna().sum()
```

Остались незаполненными данные высоты потолков, проверим что покажет функция describe()

```python
data['ceiling_height'].describe()
```

Как видим есть выбросы в датасете - врядли есть потолки с высотой 100 метров, среднее же значение вполне похоже на правду - 2,77 метра, однако стандартное отклонение составляет 1.26 метра. Взглянем поподробнее на данные где высота потолков больше 20 метров

```python
data[data['ceiling_height']>20]
```

Видим таких значений всего 22 - один выброс 100 метров и все остальные от 22 до 30 метров. Скорее всего пользователи ошиблись при вводе данных и эти значения высоты надо поделить на 10. Поделим значения больше 20 метров на 10, сохраним в датасете и посмотрим на гистрограмму

```python
data.loc[data['ceiling_height'] >= 20, 'ceiling_height'] = data['ceiling_height'] / 10
```

Посмотрим как изменились данные после исправления.

```python
data['ceiling_height'].describe()
```

Средняя стала 2.73 и стандартное отклонение уменьшилось с 1.26 до 0.31
Но как мы помним у нас еще остались пропуски в данных - посмотрим сколько их и заменим отсутствующие значения на обновленное среднее. Мы заменили некорректные значения, которые сильно выбивались (от 20 до 30 метров) но оставись значения меньше двух и больше 6 метров - с которыми непонятно что делать, посмотрим сколько их и  уберем их из расчета средней и обновленную среднюю.

```python
data[(data['ceiling_height'] < 2) | (data['ceiling_height'] > 6)]
```

```python
data[(data['ceiling_height'] > 2) & (data['ceiling_height'] < 6)]['ceiling_height'].describe()
```

## Расчёты и добавление результатов в таблицу

Добавим в таблицу цену квадратного метра квартиры

```python
data['price_per_square_meter'] = (data['last_price'] / data['total_area'])
```

Так же добавим день недели, месяц и год публикации объявления в датасет, для удобства анализа

```python
data['day_publishd'] = data['first_day_exposition'].dt.day      # день
data['month_publishd'] = data['first_day_exposition'].dt.month  # месяц
data['year_publishd'] = data['first_day_exposition'].dt.year    # год
data['weekday'] = data['first_day_exposition'].dt.dayofweek     # день недели
```

Добавим в таблицу этаж квартиры -  варианты — первый, последний, другой
Для этого создадим функцию - которой на вход будем подавать строку датасета, сравнивать этаж квартиры и этаж дома, и в зависимости от значения возвращать 3 варианта - первый, последний, другой 

```python
def floor_category (row):
    floor = row['floor']
    floor_total = row ['floors_total']
    if  floor == floor_total:
        return 'последний'
    elif floor == 1:
        return 'первый'
    else:
        return 'другой' 
```

Создадим новый столбец в датасете и применим к нему функцию floor_category

```python
data['floor_category'] = data.apply(floor_category, axis = 1)
```

Добавим в таблицу соотношение жилой и общей площади, а также отношение площади кухни к общей площади

```python
data ['living_area_ratio'] = data ['living_area'] / data['total_area'] 
data ['kitchen_area_ratio'] =  data ['kitchen_area'] / data['total_area'] 
```
